**Merge tables**

```python
data_1 = data_1.merge(data)
data_1.head()
```

```python
rb = wdg.RadioButtons(options=['Electricity','Gas','Both'], description='Usage:')
dd = wdg.Dropdown(options=list(["Briuhovychi","Gorodok","Lviv","Mostyska","Pustomyty","Sokilnyky","Stryi","Vynnyky","Yavoriv","Zhovkva"]),description='City')
btn = wdg.Button(description='Show', button_style='info', tooltip = 'Click me :)')
hb = wdg.HBox([dd,rb,btn])
```

```python
def result(b):
    if(rb.value=="Electricity"):
        return data_1[data_1.town==dd.value].groupby("date").sum()[1:].plot(y="e_meter",figsize=(12,6),title="Electricity in {0}".format(dd.value))
    elif(rb.value=="Gas"):
        return data_1[data_1.town==dd.value].groupby("date").sum()[1:].plot(y="g_meter",figsize=(12,6),title="Gas in {0}".format(dd.value))
    elif(rb.value=="Both"):
        return data_1[data_1.town==dd.value].groupby("date").sum()[1:].plot(y=["g_meter","e_meter"],figsize=(12,6),title="Gas in {0}".format(dd.value))
btn.on_click(result)
```

```python
print("Total electricity consuming=",data_1.groupby('user_id').e_meter.diff().sum())
```

```python
print("Total gas consuming=",data_1.groupby('user_id').g_meter.diff().sum())
```

```python
def e_cost(x):
    if(x.e_meter_diff <= 100):
        return x.e_meter_diff*x.upto100
    else:
        return 100*x.upto100+(x.e_meter_diff-100)*x.more
def g_cost(x):
    return x.g_meter_diff * x.gas
```

```python
rb = wdg.RadioButtons(options=['Electricity','Gas','Both'], description='Consumption:')
dd = wdg.Dropdown(options=list(data_1.user_id.unique()),description='Користувач')
btn = wdg.Button(description='Побудувати графік', button_style='info', tooltip = 'Click me :)')
hb = wdg.HBox([dd,rb,btn])
```

```python
def make_plot(b):
    if(rb.value == "Electricity"):
        data_1.groupby('user_id').get_group(dd.value)[1:].plot.bar(y="e_cost",x="date",title="Statisctics about electricity for {0}".format(dd.value),figsize=(12,6),color="r")
    elif(rb.value == "Gas"):
         data_1.groupby('user_id').get_group(dd.value)[1:].plot.bar(y="g_cost",title="Statisctics about gas for {0}".format(dd.value),figsize=(12,6),color="g")
    elif(rb.value=="Both"):
        data_1.groupby('user_id').get_group(dd.value)[1:].plot.bar(y=["e_cost",'g_cost'],title="Statisctics about gas and electricity for {0}".format(dd.value),figsize=(12,6))
btn.on_click(make_plot)
```
